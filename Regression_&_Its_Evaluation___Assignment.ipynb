Question 1:  What is Simple Linear Regression ?

Simple Linear Regression is a statistical method used to model the relationship between two continuous variables: a dependent variable (the one you want to predict) and an independent variable (the one you use to predict). It assumes a linear relationship, meaning the relationship can be represented by a straight line. The goal is to find the best-fitting line that minimizes the distance between the observed data points and the line. This line is defined by two parameters: the slope and the intercept. The slope indicates how much the dependent variable changes for every unit change in the independent variable, and the intercept is the value of the dependent variable when the independent variable is zero.

Question 2: What are the key assumptions of Simple Linear Regression ?

The key assumptions of Simple Linear Regression are:

1.  **Linearity:** The relationship between the independent and dependent variables is linear.
2.  **Independence:** The observations are independent of each other.
3.  **Homoscedasticity:** The variance of the residuals (the differences between the observed and predicted values) is constant across all levels of the independent variable.
4.  **Normality:** The residuals are normally distributed.
5.  **No multicollinearity:** If there are multiple independent variables (in the case of multiple linear regression), they should not be highly correlated with each other. (While this is for multiple linear regression, it's still a good point to mention for completeness).

Question 3: What is heteroscedasticity, and why is it important to address in regression models?

**Heteroscedasticity** occurs when the variance of the residuals (the errors or differences between the observed and predicted values) in a regression model is not constant across all levels of the independent variable(s). In simpler terms, the spread of the data points around the regression line is not consistent; it might be wider for some values of the independent variable and narrower for others.

It is important to address heteroscedasticity in regression models for several reasons:

1.  **Inefficient Coefficient Estimates:** While the coefficient estimates in the presence of heteroscedasticity are still unbiased, they are no longer the most efficient (they have larger standard errors). This means the estimates are less precise.
2.  **Incorrect Standard Errors:** The standard errors of the regression coefficients will be biased (usually underestimated) when heteroscedasticity is present.
3.  **Invalid Statistical Inference:** Because the standard errors are incorrect, hypothesis tests (like t-tests) and confidence intervals will be unreliable. This can lead to incorrect conclusions about the statistical significance of the independent variables.
4.  **Misleading Model Fit:** Measures of model fit, such as R-squared, can be misleading when heteroscedasticity is present.

Addressing heteroscedasticity can involve various techniques, such as:

*   Transforming the dependent variable (e.g., using logarithms or square roots).
*   Using weighted least squares regression, which gives less weight to observations with larger variances.
*   Using robust standard errors, which provide correct standard errors even in the presence of heteroscedasticity.

Question 4: What is Multiple Linear Regression ?

Multiple Linear Regression is an extension of simple linear regression used to model the relationship between a single dependent variable and **two or more** independent variables. Like simple linear regression, it assumes a linear relationship between the dependent variable and each of the independent variables. The goal is to find the best-fitting linear combination of the independent variables that minimizes the distance between the observed data points and the predicted values. The model estimates coefficients for each independent variable, which represent the change in the dependent variable associated with a one-unit change in that independent variable, holding all other independent variables constant.

Question 5: What is polynomial regression, and how does it differ from linear
regression?  


**Polynomial Regression** is a form of regression analysis in which the relationship between the independent variable \(x\) and the dependent variable \(y\) is modeled as an n-th degree polynomial. While linear regression models a straight-line relationship, polynomial regression allows for a curved relationship.

The key difference between polynomial regression and linear regression lies in the form of the equation:

*   **Linear Regression:** \(y = \beta_0 + \beta_1x + \epsilon\)
*   **Polynomial Regression:** \(y = \beta_0 + \beta_1x + \beta_2x^2 + ... + \beta_nx^n + \epsilon\)

Where:
*   \(y\) is the dependent variable.
*   \(x\) is the independent variable.
*   \(\beta_0\) is the intercept.
*   \(\beta_1, \beta_2, ..., \beta_n\) are the coefficients for the polynomial terms.
*   \(\epsilon\) is the error term.

In essence, polynomial regression introduces non-linear terms (like \(x^2\), \(x^3\), etc.) into the linear model equation. This allows the model to capture more complex, non-linear patterns in the data that a simple linear regression cannot.

While polynomial regression is still considered a form of linear model because the coefficients (\(\beta\)) are linear, the relationship between \(x\) and \(y\) is non-linear. This makes it more flexible than simple linear regression for modeling data that doesn't follow a straight line.

Question 6:  Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Predict Y values using the fitted model
Y_pred = model.predict(X)

# Plot the original data points and the regression line
plt.scatter(X, Y, color='blue', label='Original data')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Print the coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficient (slope): {model.coef_[0]}")
```

Question 8:  Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]

● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (2nd degree)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

# Create a linear regression model (this will fit the polynomial features)
model = LinearRegression()

# Fit the model to the polynomial features
model.fit(X_poly, Y)

# Predict Y values using the fitted model
Y_pred = model.predict(X_poly)

# Plot the original data points
plt.scatter(X, Y, color='blue', label='Original data')

# Plot the polynomial regression curve
# To get a smooth curve, we need more points for the prediction
X_smooth = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_smooth_poly = poly_features.transform(X_smooth)
Y_smooth_pred = model.predict(X_smooth_poly)
plt.plot(X_smooth, Y_smooth_pred, color='red', linewidth=2, label='Polynomial Regression (Degree 2)')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.grid(True)
plt.show()

# Print the coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

Question 9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Predict Y values using the fitted model
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Create a residuals plot
plt.scatter(X, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--') # Add a horizontal line at y=0
plt.xlabel('X')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()

# Assess heteroscedasticity by examining the spread of residuals
print("Assessing heteroscedasticity:")
print("Observe the residuals plot. If the spread of residuals appears to be roughly constant across all values of X, then homoscedasticity is likely present.")
print("If the spread of residuals widens or narrows as X increases, this suggests heteroscedasticity.")
```

Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

This is a common and important scenario in real-world data science! Addressing heteroscedasticity and multicollinearity is crucial for building a reliable and robust regression model. Here are the steps I would take as a data scientist in this situation:

**Addressing Heteroscedasticity:**

1.  **Confirm Heteroscedasticity:** I would first visually confirm heteroscedasticity using a residuals plot (plotting residuals against predicted values or independent variables). Statistical tests like the Breusch-Pagan test or White test could also be used for a more formal assessment.

2.  **Transformations:** I would consider transforming the dependent variable (house price) or the independent variables. Common transformations include:
    *   **Log Transformation:** Taking the logarithm of the house price is often effective as it can compress the range of larger values and stabilize the variance.
    *   **Square Root or Cube Root Transformations:** These can also help in reducing the spread of residuals for larger values.

3.  **Weighted Least Squares (WLS):** If transformations are not effective or appropriate, I would consider using Weighted Least Squares regression. WLS assigns lower weights to observations with higher variances (where the spread of residuals is larger) and higher weights to observations with lower variances. This gives more influence to the data points that are more precisely measured.

4.  **Robust Standard Errors:** Even if the heteroscedasticity is not fully eliminated, using robust standard errors (also known as heteroscedasticity-consistent standard errors) can provide correct standard errors for the regression coefficients. This allows for valid statistical inference (hypothesis testing and confidence intervals) even in the presence of heteroscedasticity.

**Addressing Multicollinearity:**

1.  **Identify Multicollinearity:** I would identify multicollinearity by examining the correlation matrix of the independent variables. High correlation coefficients (e.g., above 0.8) between two or more independent variables indicate potential multicollinearity. Variance Inflation Factor (VIF) is a more formal measure; VIF values above 5 or 10 are often considered problematic.

2.  **Variable Selection:**
    *   **Remove Highly Correlated Variables:** If two or more variables are highly correlated, I would consider removing one or more of them. The decision of which variable to remove might be based on domain knowledge (which variable is more theoretically relevant), practical considerations (which variable is easier to collect or interpret), or statistical significance in preliminary models.
    *   **Feature Engineering:** Sometimes, multicollinearity can be addressed by combining highly correlated variables into a single new feature. For example, if "number of bedrooms" and "number of bathrooms" are highly correlated, a new feature representing "total number of rooms" might be created.

3.  **Dimensionality Reduction Techniques:** Techniques like Principal Component Analysis (PCA) can be used to create a new set of uncorrelated variables (principal components) from the original independent variables. The regression can then be performed on these principal components.

4.  **Regularization Techniques:** Ridge Regression and Lasso Regression are regularization techniques that can help mitigate the impact of multicollinearity. They add a penalty term to the regression objective function that shrinks the coefficients, particularly for correlated variables. Lasso can also perform variable selection by shrinking some coefficients to zero.

**Ensuring a Robust Model:**

After addressing heteroscedasticity and multicollinearity, I would:

1.  **Re-evaluate Assumptions:** I would re-check the assumptions of the regression model (linearity, normality of residuals, etc.) using the residuals plot and other diagnostic tools.

2.  **Model Evaluation:** I would evaluate the model's performance using appropriate metrics (e.g., R-squared, Adjusted R-squared, Mean Squared Error, Root Mean Squared Error) on a separate validation or test set to ensure it generalizes well to new data.

3.  **Interpretability:** I would ensure that the model is interpretable and the coefficients make sense in the context of the real estate market.
